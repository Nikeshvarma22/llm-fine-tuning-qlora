```python
CATEGORIES = ["billing", "technical_support", "account_access", "product_inquiry", "cancellation"]

# Sample tickets for each category — these simulate real customer support data
ticket_templates = {
    "billing": [
        "I was charged twice for my subscription this month. Please refund the extra payment.",
        "My credit card was billed $49.99 but my plan is only $29.99. What happened?",
        "I need a copy of my invoice from last month for tax purposes.",
        "Can you explain the charges on my latest bill? I see fees I don't recognize.",
        "I upgraded my plan but I'm still being charged the old rate.",
        "When will my refund be processed? It's been 5 business days already.",
        "I cancelled last week but still got charged today. This is unacceptable.",
        "My payment failed but the service is still active. Will I be charged later?",
        "I want to switch from annual to monthly billing. How does that affect my rate?",
        "There's a pending charge on my account that I didn't authorize.",
        "Can I get a discount if I pay for the full year upfront?",
        "My promo code didn't apply correctly and I was charged full price.",
        "I need to update my billing address for tax compliance.",
        "Why was I charged a late fee? I set up auto-pay last month.",
        "I disputed a charge with my bank. Will my account be affected?",
        "The currency conversion on my bill seems wrong. I'm in India, charged in USD.",
        "Can I split my payment across two different credit cards?",
        "I received a bill for a service I never signed up for.",
        "My company needs a consolidated invoice for all our team accounts.",
        "The billing cycle changed without any notification. Why?",
        "I was promised a 30-day free trial but got charged on day 15.",
        "How do I download my billing history for the past year?",
        "My student discount expired but I'm still enrolled. Can you reapply it?",
        "I see a charge labeled 'premium add-on' but I never added any extras.",
        "Can I get a refund for the unused portion of my annual subscription?",
    ],
    "technical_support": [
        "The app keeps crashing whenever I try to open the dashboard.",
        "I'm getting a 504 gateway timeout error when loading reports.",
        "My file uploads are failing with a 'size limit exceeded' error but the file is only 2MB.",
        "The search function isn't returning any results even though I know the data exists.",
        "API responses are taking over 30 seconds. Is there a performance issue?",
        "I can't connect the integration with Slack. It shows an OAuth error.",
        "The export to CSV feature is generating corrupted files.",
        "My notifications stopped working two days ago. I'm not getting any alerts.",
        "The mobile app shows a blank white screen after the latest update.",
        "I'm getting a 'session expired' error every 5 minutes even though I'm active.",
        "The drag and drop feature doesn't work on my Chrome browser.",
        "Images I upload appear rotated 90 degrees in the preview.",
        "The real-time sync between devices isn't working. Changes don't appear.",
        "I keep getting 'insufficient permissions' error but I'm an admin.",
        "The dark mode toggle doesn't save my preference. It resets every time.",
        "PDF generation is failing with a memory allocation error.",
        "The calendar widget shows wrong timezone. I'm in IST but it shows PST.",
        "Auto-save isn't working. I lost 2 hours of work when my browser crashed.",
        "The two-factor authentication code isn't being accepted even though it's correct.",
        "Loading speed has degraded significantly in the past week.",
        "The webhook endpoint returns 400 Bad Request for valid payloads.",
        "My custom CSS isn't being applied to the embedded widget.",
        "The search filters reset when I switch between pages.",
        "Batch operations time out when processing more than 100 items.",
        "The SSL certificate warning appears when accessing the API endpoint.",
    ],
    "account_access": [
        "I forgot my password and the reset email isn't arriving.",
        "My account has been locked after too many login attempts.",
        "I need to change the email address associated with my account.",
        "Can I merge my two accounts into one? I accidentally created a duplicate.",
        "I'm being asked for two-factor authentication but I lost my phone.",
        "My team member needs admin access but I can't find where to change roles.",
        "I haven't received the email verification link to activate my account.",
        "I'm locked out because my company changed the SSO provider.",
        "How do I transfer account ownership to my colleague?",
        "My account shows 'suspended' but I haven't violated any terms.",
        "I need to recover my account. I no longer have access to my old email.",
        "Can I sign in with Google instead of email and password?",
        "The login page keeps redirecting me in a loop. I can never get in.",
        "I want to enable single sign-on for my entire organization.",
        "My temporary password expired before I could log in.",
        "I need access to my account data but I deleted my account last month.",
        "The security question I set doesn't match what the system is asking.",
        "I'm getting 'this account doesn't exist' but I've been a customer for 2 years.",
        "Can I add a backup email address in case I lose access to my primary one?",
        "My multi-factor authentication app was reset and I can't generate codes.",
        "I need to remove a former employee's access from our team account.",
        "The passwordless login link expires too quickly. I can't click it in time.",
        "I want to set up different permission levels for my team members.",
        "My account was compromised. Someone changed my email and password.",
        "I can't accept the team invitation. The link says it's already been used.",
    ],
    "product_inquiry": [
        "What features are included in the Pro plan vs the Basic plan?",
        "Do you offer an API for integrating with our internal tools?",
        "Can your platform handle 10,000 concurrent users?",
        "Is there a mobile app available for iOS and Android?",
        "What programming languages does your SDK support?",
        "Do you have a feature for automated reporting?",
        "Can I white-label the product for my clients?",
        "What's your uptime SLA for the enterprise plan?",
        "Do you support data export in JSON and XML formats?",
        "Is there a sandbox environment for testing before going live?",
        "What's the maximum file size I can upload?",
        "Do you offer on-premise deployment or is it cloud-only?",
        "Can I schedule automated backups of my data?",
        "What compliance certifications does your platform have? SOC2? GDPR?",
        "Is there a bulk import feature for migrating from another platform?",
        "Do you support custom fields and workflows?",
        "What analytics and reporting dashboards are available?",
        "Can I connect your platform with Salesforce and HubSpot?",
        "Do you have multi-language support for international teams?",
        "What's the difference between your team and enterprise plans?",
        "Is there a self-hosted option for companies with strict data policies?",
        "Do you provide an AI assistant feature in any of your plans?",
        "Can I create custom roles and permissions for my organization?",
        "What's your data retention policy? How long is data stored?",
        "Do you offer a free trial before committing to a paid plan?",
    ],
    "cancellation": [
        "I want to cancel my subscription immediately.",
        "How do I cancel? I can't find the option in settings.",
        "I'm cancelling because the price increase is too much.",
        "Will I lose all my data if I cancel my account?",
        "I want to cancel but keep access until the end of my billing period.",
        "I'm switching to a competitor. Please cancel and export my data.",
        "Can I pause my subscription instead of cancelling? I might come back.",
        "I need to cancel because our company is switching to a different solution.",
        "Please cancel my account and confirm via email.",
        "I'm not using the service enough to justify the cost. Cancel please.",
        "I want to downgrade to the free plan instead of fully cancelling.",
        "What happens to my integrations if I cancel?",
        "I'd like to cancel my annual plan and get a prorated refund.",
        "Cancel my subscription. The features promised during sales were misleading.",
        "I've been trying to cancel for a week but the button doesn't work.",
        "Is there an exit survey? I want to provide feedback before cancelling.",
        "I'm cancelling all 5 seats on my team plan. How do I do that?",
        "Can you explain the cancellation policy? Is there an early termination fee?",
        "I'm cancelling because the customer support response time is too slow.",
        "I want to cancel but I need a data export first. How do I get that?",
        "Please confirm that no further charges will be made after cancellation.",
        "I'm unhappy with the recent UI changes. Cancelling my subscription.",
        "My free trial is ending and I don't want to be charged. Cancel it.",
        "I want to delete my account entirely, not just cancel the subscription.",
        "Is there a retention offer? I might stay if you can match the competitor's price.",
    ],
}

def create_fine_tuning_dataset():
    """
    Create train and test datasets in instruction format.

    INSTRUCTION FORMAT — the standard for fine-tuning modern LLMs:

    <s>[INST] Classify this customer support ticket into one of these categories:
    billing, technical_support, account_access, product_inquiry, cancellation.

    Ticket: "I was charged twice this month"

    Respond with only the category name. [/INST] billing</s>

    WHY THIS FORMAT?
    - [INST]...[/INST] tells the model "this is the instruction"
    - The text after [/INST] is what we want the model to learn to generate
    - <s> and </s> are start/end of sequence tokens
    - This is the format Mistral was pre-trained with, so fine-tuning works best
    """

    all_examples = []

    for category, tickets in ticket_templates.items():
        for ticket in tickets:
            # Create the instruction-formatted example
            text = f"""<s>[INST] Classify this customer support ticket into one of these categories: billing, technical_support, account_access, product_inquiry, cancellation.

Ticket: "{ticket}"

Respond with only the category name. [/INST] {category}</s>"""

            all_examples.append({
                "text": text,
                "category": category,
                "ticket": ticket,
            })

    # Shuffle
    random.seed(42)  # Reproducibility — same shuffle every time
    random.shuffle(all_examples)

    # Split: 80% train, 20% test
    split_idx = int(len(all_examples) * 0.8)
    train_data = all_examples[:split_idx]
    test_data = all_examples[split_idx:]

    train_dataset = Dataset.from_list(train_data)
    test_dataset = Dataset.from_list(test_data)

    print(f"✅ Dataset created!")
    print(f"   Training examples: {len(train_data)}")
    print(f"   Test examples: {len(test_data)}")
    print(f"   Categories: {CATEGORIES}")
    print(f"\n📝 Sample training example:\n")
    print(train_data[0]["text"][:300] + "...")

    return train_dataset, test_dataset

# Create the datasets
train_dataset, test_dataset = create_fine_tuning_dataset()
```

```python
model_name = "mistralai/Mistral-7B-Instruct-v0.3"

# ── Step 1: Configure 4-bit quantization ──────────────────
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,                    # Load in 4-bit precision
    bnb_4bit_quant_type="nf4",            # NormalFloat4 — best for LLMs
    bnb_4bit_compute_dtype=torch.float16, # Compute in FP16 for speed
    bnb_4bit_use_double_quant=True,       # Quantize the quantization constants (saves more memory)
)

print("📥 Loading Mistral-7B in 4-bit... (this takes 2-3 minutes)")

# ── Step 2: Load tokenizer ────────────────────────────────
# The tokenizer converts text ↔ numbers
# "Hello world" → [22557, 1526] → model processes numbers → [8876, 442] → "Fine, thanks"
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token  # Mistral doesn't have a pad token by default
tokenizer.padding_side = "right"           # Pad on the right side

# ── Step 3: Load model in 4-bit ───────────────────────────
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",         # Automatically place layers on GPU
    trust_remote_code=True,
)

# Disable caching (not compatible with gradient checkpointing during training)
model.config.use_cache = False

print(f"\n✅ Model loaded!")
print(f"   Model: {model_name}")
print(f"   Parameters: ~7 Billion")
print(f"   Loaded in: 4-bit quantization")
print(f"   GPU Memory used: {torch.cuda.memory_allocated() / 1e9:.1f} GB")
```

```python
def classify_with_model(model, tokenizer, ticket_text, max_new_tokens=20):
    """
    Send a ticket to the model and get its classification.
    Returns the predicted category.
    """
    prompt = f"""<s>[INST] Classify this customer support ticket into one of these categories: billing, technical_support, account_access, product_inquiry, cancellation.

Ticket: "{ticket_text}"

Respond with only the category name. [/INST]"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():  # Don't compute gradients (we're just predicting)
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            temperature=0.1,      # Low temp = more deterministic
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id,
        )

    # Decode only the NEW tokens (not the prompt)
    response = tokenizer.decode(outputs[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)

    # Extract category from response
    response_lower = response.strip().lower()
    for cat in CATEGORIES:
        if cat in response_lower:
            return cat

    return response_lower.strip().split('\n')[0].strip()  # Return first line if no match

def evaluate_model(model, tokenizer, test_dataset, label="Model"):
    """
    Run the model on all test examples and compute accuracy.
    """
    predictions = []
    true_labels = []

    print(f"\n🧪 Evaluating {label} on {len(test_dataset)} test examples...")

    for i, example in enumerate(test_dataset):
        if i % 5 == 0:
            print(f"   Processing {i+1}/{len(test_dataset)}...")

        pred = classify_with_model(model, tokenizer, example["ticket"])
        predictions.append(pred)
        true_labels.append(example["category"])

    # Calculate accuracy
    # Normalize predictions to match category names
    normalized_preds = []
    for pred in predictions:
        matched = False
        for cat in CATEGORIES:
            if cat in pred.lower():
                normalized_preds.append(cat)
                matched = True
                break
        if not matched:
            normalized_preds.append(pred)

    accuracy = accuracy_score(true_labels, normalized_preds)

    print(f"\n{'='*50}")
    print(f"📊 {label} Results:")
    print(f"   Accuracy: {accuracy:.1%}")
    print(f"{'='*50}")

    # Show some example predictions
    print(f"\n📝 Sample predictions:")
    for i in range(min(5, len(test_dataset))):
        status = "✅" if normalized_preds[i] == true_labels[i] else "❌"
        print(f"   {status} Ticket: \"{test_dataset[i]['ticket'][:60]}...\"")
        print(f"      True: {true_labels[i]} | Predicted: {normalized_preds[i]}")

    return accuracy, normalized_preds, true_labels

# Evaluate BASE model (before fine-tuning)
print("📊 Testing BASE model (before fine-tuning)...")
print("   This shows how the model performs WITHOUT our training.\n")
base_accuracy, base_preds, true_labels = evaluate_model(
    model, tokenizer, test_dataset, label="BASE Model (Before Fine-Tuning)"
)
```

```python
model = prepare_model_for_kbit_training(model)

# Configure LoRA
lora_config = LoraConfig(
    r=16,                          # Rank — 16 is the sweet spot for most tasks
    lora_alpha=32,                 # Scaling factor (usually 2× the rank)
    lora_dropout=0.05,             # Small dropout to prevent overfitting
    bias="none",                   # Don't train bias terms
    task_type="CAUSAL_LM",        # We're fine-tuning a causal language model
    target_modules=[               # WHICH layers to add LoRA to
        "q_proj",                  # Query projection (attention)
        "k_proj",                  # Key projection (attention)
        "v_proj",                  # Value projection (attention)
        "o_proj",                  # Output projection (attention)
    ],
)

# Wrap the model with LoRA adapters
model = get_peft_model(model, lora_config)

# Print trainable parameters
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
total_params = sum(p.numel() for p in model.parameters())
print(f"\n✅ LoRA configured!")
print(f"   Total parameters:     {total_params:,}")
print(f"   Trainable parameters: {trainable_params:,}")
print(f"   Trainable %:          {100 * trainable_params / total_params:.2f}%")
print(f"   Rank (r):             {lora_config.r}")
print(f"   Target modules:       {lora_config.target_modules}")
```

```python
training_args = TrainingArguments(
    output_dir="./fine_tuned_model",

    # ── How much to train ──
    num_train_epochs=3,               # Go through the data 3 times

    # ── Batch size & memory ──
    per_device_train_batch_size=4,    # 4 examples per GPU step
    gradient_accumulation_steps=4,    # Accumulate 4 steps → effective batch = 16
    # WHY accumulate? T4 can't fit batch_size=16 in memory.
    # So we do 4 steps of 4, accumulate gradients, then update.
    # Same effect as batch_size=16 but uses less memory.

    # ── Learning rate ──
    learning_rate=2e-4,               # 0.0002 — standard for LoRA fine-tuning
    weight_decay=0.01,                # Small regularization
    warmup_ratio=0.03,                # Warm up for first 3% of steps
    # WHY warmup? Starting with a high learning rate can destabilize
    # a quantized model. Warmup gradually increases the LR.

    # ── Optimization ──
    optim="paged_adamw_8bit",         # 8-bit Adam optimizer (saves memory)
    fp16=False,
    bf16=True,                       # Use FP16 for speed

    # ── Logging ──
    logging_steps=10,                 # Log every 10 steps
    save_strategy="epoch",            # Save after each epoch

    # ── Other ──
    gradient_checkpointing=True,      # Trade compute for memory
    max_grad_norm=0.3,                # Clip gradients (prevents exploding gradients)
    report_to="none",                 # Don't log to wandb
)

trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    args=training_args,
    processing_class=tokenizer,
)

print("🚀 Starting fine-tuning...")
print(f"   Epochs: {training_args.num_train_epochs}")
print(f"   Effective batch size: {training_args.per_device_train_batch_size * training_args.gradient_accumulation_steps}")
print(f"   Learning rate: {training_args.learning_rate}")
print(f"   This will take ~15-20 minutes on T4 GPU\n")

# TRAIN!
train_result = trainer.train()

print(f"\n✅ Training complete!")
print(train_result.metrics)
print(f"   Total steps: {train_result.metrics['train_steps']}")
print(f"   Training loss: {train_result.metrics['train_loss']:.4f}")
print(f"   Training time: {train_result.metrics['train_runtime']:.0f} seconds")

# Save the fine-tuned model
trainer.save_model("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")
print(f"\n💾 Model saved to ./fine_tuned_model")
```

```python
print("📊 Testing FINE-TUNED model...")
ft_accuracy, ft_preds, true_labels = evaluate_model(
    model, tokenizer, test_dataset, label="FINE-TUNED Model (After Training)"
)

# ── Compare Before vs After ──
print(f"\n{'='*50}")
print(f"📈 IMPROVEMENT SUMMARY")
print(f"{'='*50}")
print(f"   Base model accuracy:       {base_accuracy:.1%}")
print(f"   Fine-tuned accuracy:       {ft_accuracy:.1%}")
print(f"   Improvement:               +{(ft_accuracy - base_accuracy):.1%}")
print(f"{'='*50}")

# ── Detailed Classification Report ──
print(f"\n📋 Detailed Classification Report:\n")
print(classification_report(true_labels, ft_preds, labels=CATEGORIES, zero_division=0))
```
